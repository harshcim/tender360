```python
import subprocess
import os
```

```python
pdf_path = '/home/cimcon/Desktop/pdfs_for_classification /relevant/Nandurbar3RDCALL.pdf'
page_number = 56
```

```python
# Open the PDF file to a specific page (using default PDF viewer)
if os.name == 'nt':  # For Windows
    os.startfile(f'{pdf_path}#page={page_number}')
elif os.name == 'posix':  # For MacOS and Linux
    subprocess.run(['open', f'{pdf_path}#page={page_number}'])
else:
    print("Unsupported OS!")
```
